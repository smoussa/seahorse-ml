```python
#Read in all of the trips from data files 1.csv to 200.csv into a list
#Each element of the list named "data" is NumPy array of x, y coordinates
data = [np.genfromtxt(str(i) +".csv", skip_header=1, delimiter=",") for i in range(1, 201)]
```

```python
#Since the trips have been transformed to start at the origin, absolute positions don't really matter
#We are instead concerned with differences between each data point
data_dt = [np.diff(datum, axis=0) for datum in data]
```

```python
def find_angle(v):
    if np.linalg.norm(v) == 0:
        return 0
    xaxis = [1, 0]
    cos_sim = np.dot(xaxis, v)/np.linalg.norm(v)
    return np.arccos(cos_sim)

#Obtain the angles at which the car is travelling for each of the trips

def no_zeros(a):
    for i, item in enumerate(a):
        if a[i] == 0:
            a[i] = a[i - 1]

thetas = [np.apply_along_axis(find_angle, 1, deltas) for deltas in data_dt]
for t in thetas:
    no_zeros(t)
```

```python
#Because trips are randomly rotated only the change in direction really matters
#So here we take the differences of the angles
thetas_dt = [np.diff(theta) for theta in thetas]
```
